```python
from PIL import Image, ImageDraw
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
from ctypes import sizeof
colors = [(66, 30, 15),     # brown 3
          (25, 7, 26),      # dark violett
          (9, 1, 47),       # darkest blue
          (4, 4, 73),       # blue 5
          (0, 7, 100),      # blue 4
          (12, 44, 138),    # blue 3
          (24, 82, 177),    # blue 2
          (57, 125, 209),   # blue 1
          (134, 181, 229),  # blue 0
          (211, 236, 248),  # lightest blue
          (241, 233, 191),  # lightest yellow
          (248, 201, 95),   # light yellow
          (255, 170, 0),    # dirty yellow
          (204, 128, 0),    # brown 0
          (153, 87, 0),     # brown 1
          (106, 52, 3)]     # brown 2
imagesize = 1000

def draw_image_from_txt(txt_file, output_image, mode=0):
    # read text files
    with open(txt_file, 'r') as file:
        # read each line of the file
        lines = file.readlines()

    # determines the size of the image based on the number of rows and elements per row
    height = len(lines)
    width = max(len(line.strip().split(',')) for line in lines)

    # create a blank white image with the appropriate size
    image = Image.new('RGB', (width, height), color='white')

    # for each line of the
    for y, line in enumerate(lines):
        # remove any whitespace and newlines from the line
        line = line.strip()

        # split the line using commas as separator
        values = line.split(',')

        # for each element in the row
        for x, value in enumerate(values):
            # converts value to integer
            pixel_value = int(value)

            # assigns the pixel color based on the read value
            if mode == 0:
                color = (pixel_value, pixel_value, pixel_value)  # grayscale
            else:
                if pixel_value == 0:
                    color = (0, 0, 0)
                else:
                    color = colors[pixel_value % len(colors)]

            # sets the color of the pixel in the image
            image.putpixel((x, y), color)

    # save the resulting image
    image.save(output_image)
    img = mpimg.imread(output_image)
    plt.imshow(img)
    plt.axis('off')  # Hide the axes
    plt.show()

    print(f"L'immagine {output_image} Ã¨ stata salvata.")
```

```python
input_txt_file = "output4.txt"  # Nome del file di testo contenente le coordinate
output_image_file1 = "output_grigi.png"  # Nome del file immagine di output (scala di grigi)
output_image_file2 = "output_setcolori.png"  # Nome del file immagine di output (set di colori)
output_image_file3 = "output_scale.png"  # Nome del file immagine di output (scale colorate)

draw_image_from_txt(input_txt_file, output_image_file1, 0)
draw_image_from_txt(input_txt_file, output_image_file2, 1)
#draw_image_from_txt(input_txt_file, output_image_file3, 2)
```
